# Data Visualization

# Closing price of various Cryptocurrency in 2018 vs 2020

```python

# Ensure 'Date' columns are in datetime format
for df in [btc, ETH, bitconnect, doge]:
    df['Date'] = pd.to_datetime(df['Date'])

# Function to filter data for the year 2018
def to2018(df):
    return df[df['Date'].dt.year == 2018]

# Filtered DataFrames for 2018
btc18 = to2018(btc)
eth18 = to2018(ETH)
bit18 = to2018(bitconnect)

# List of DataFrames and their corresponding titles
crypto_data = [btc18, btc, eth18, ETH, bit18, doge]
crypto_titles = ["Bitcoin 2018", "Bitcoin", "Ethereum 2018", "Ethereum", "Bitconnect 2018", "Dogecoin"]

# Plot settings
num_plots = len(crypto_data)
cols = 2
rows = (num_plots + 1) // cols  # Ensure enough rows

fig, axs = plt.subplots(nrows=rows, ncols=cols, figsize=(14 * cols, 6 * rows), constrained_layout=True)
axs = axs.flatten()  # Flatten in case of single row

sns.set_context('paper', font_scale=1.5)

for i, (df, title) in enumerate(zip(crypto_data, crypto_titles)):
    sns.lineplot(data=df, x="Date", y="Close", ax=axs[i], color='#732C2C')
    axs[i].set_title(title)
    axs[i].tick_params(axis='x', rotation=45)

# Hide any unused subplots
for j in range(i + 1, len(axs)):
    fig.delaxes(axs[j])

plt.show()
```

```python
def equalize(df1, df2):
    min_len = min(len(df1), len(df2))
    return df1.tail(min_len).reset_index(drop=True)
    
# Equalize lengths based on ETH's index
btc = equalize(btc, ETH)
doge = equalize(doge, ETH)

# Create DataFrame of aligned closing prices
closeDf = pd.DataFrame({
    'BTC': btc['Close'].reset_index(drop=True),
    'ETH': ETH['Close'].reset_index(drop=True),
    'DOGE': doge['Close'].reset_index(drop=True)
})

# Calculate daily percentage returns
returns = closeDf.pct_change()

# Preview the result
returns.head()
```

```python

from matplotlib.colors import LinearSegmentedColormap

# Define custom colormap function
def NonLinCdict(steps, hexcol_array):
    cdict = {'red': (), 'green': (), 'blue': ()}
    for s, hexcol in zip(steps, hexcol_array):
        rgb = matplotlib.colors.hex2color(hexcol)
        cdict['red'] += ((s, rgb[0], rgb[0]),)
        cdict['green'] += ((s, rgb[1], rgb[1]),)
        cdict['blue'] += ((s, rgb[2], rgb[2]),)
    return cdict

# Define color stops and hex colors
hex_colors = ['#F8EDED', '#EAC8C8', '#CF7F7F', '#BA4949', '#732C2C']
thresholds = [0, 0.1, 0.5, 0.9, 1]

# Create custom colormap
cdict = NonLinCdict(thresholds, hex_colors)
custom_cmap = LinearSegmentedColormap('custom_cmap', cdict)

# Add DOGE to returns18 if not already there
returns18['DOGE'] = doge['Close'].pct_change().reset_index(drop=True)

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.set_context("notebook", font_scale=1.2)
sns.heatmap(returns18.corr(), annot=True, cmap=custom_cmap, fmt=".2f", linewidths=0.5)

plt.title("Correlation of Daily Returns (2018)", fontsize=14)
plt.tight_layout()
plt.show()
```
